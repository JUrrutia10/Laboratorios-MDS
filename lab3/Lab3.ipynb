### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempe√±o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a√±o y los que han tenido desempe√±o de destacado. Para esto se considerar√° que un alumno ha reprobado el a√±o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar√° destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci√≥n que permita obtener tres Dataframes:
- El primero deber√° poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser√° un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a√±o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
  reprobados = dataframe_in[dataframe_in['GPA'] < 4.0]

  df_reprobados = dataframe_in.copy()
  df_reprobados.loc[df_reprobados.index.isin(reprobados.index), 'GPA'] = 'R'

  destacados = dataframe_in[(dataframe_in['math score'] > 5.8) & (dataframe_in['reading score'] > 5.8) & (dataframe_in['writing score'] > 5.8)].sort_values(by=['GPA'], ascending=False)

  reprobados = reprobados.reset_index(drop=True)
  destacados = destacados.reset_index(drop=True)

  return reprobados, df_reprobados, destacados
````

### 3.4 Quintiles de Rendimiento [0.5 Puntos]

El rector, intrigado por saber si existe una relaci√≥n entre el nivel educacional de los padres y las notas de sus hijos les plantea estudiar dicha relaci√≥n. Para esto, les solicita que desde el `DataFrame` que posee a todos los alumnos (`DataFrame` obtenido tras la limpieza), generen una nueva columna que clasifique en quintiles de rendimiento para sus alumnos: `['Baj√≠simo', 'Bajo', 'Medio', 'Alto', 'Muy alto']`. (investigar `qcut()`)

Hecho esto, grafique a trav√©s de un scatter plot el nivel educacional de los padres v/s el promedio de los alumnos, utilice los cuantiles generados anteriormente para mostrar en que cuantil se encuentra cada uno de los datos.

Realizado el grafico, responda la siguiente pregunta: **¬øEs posible visualizar alguna tendencia en el gr√°fico?.**



**Respuesta:**

````python
# C√≥digo para quintiles


"""
NOTA: La variable del dataframe que contenga los quintiles debe llamarse
'GPA cuantiles' para que el gr√°fico funcione.
"""

def clasificar_quintiles(dataframe_in):
  dataframe_in['GPA cuantiles'] = pd.qcut(dataframe_in['GPA'], q=5, labels=['Baj√≠simo', 'Bajo', 'Medio', 'Alto', 'Muy alto'])
  return dataframe_in
````

````python
df_grades = clasificar_quintiles(df_grades)
````

````python
# C√≥digo de gr√°fico del amigo ChatGPT
fig = px.scatter(
    df_grades.sort_values(by="GPA"),
    x="GPA",
    y="parental level of education",
    hover_data=["gender"],
    color="GPA cuantiles",
    title="GPA por Nivel de educaci√≥n de los padres",
    color_discrete_sequence=["tomato", "orange", "yellow", "greenyellow", "green"]
)

fig
````

Al visualizar los cuantiles m√°s altos en todos los niveles, realmente no se nota ninguna tendencia. Pero si se observan los m√°s bajos, hay una tendencia de que mientras mayor es el nivel de educaci√≥n de los padres, m√°s alto es el GPA.

````python
df_other_grades =pd.read_csv('/content/drive/MyDrive/Lab programacioÃÅn cientifica/lab3/other_grades.csv')
display(df_other_grades)
df_grades =pd.read_json ('/content/drive/MyDrive/Lab programacioÃÅn cientifica/lab3/students_grades.json')
numeric_cols = ['math score', 'reading score', 'writing score']
df_grades[numeric_cols] = df_grades[numeric_cols].apply(pd.to_numeric, errors='coerce')
display(df_grades)
````

````python
outer_merge = pd.merge(df_grades, df_other_grades, how='outer', on='names')
display(outer_merge)
````

El outer merge combina todos los elementos posibles y conserva todo el resto en filas independientes.  Incluye todas las filas de ambos DataFrames, completando con NaNs donde no haya coincidencia entre las columnas 'names'.

````python
left_merge = pd.merge(df_grades, df_other_grades, how='left', on='names')
display(left_merge)
````

El left merge conserva solo los elementos que se hayan combinado correctamente provenientes dataset `left`, en este caso `df_grades`. Incluye todas las filas de `df_grades`, completando con NaNs donde no haya coincidencia entre las columnas 'name' en `other_grades.csv`. Este tipo de uni√≥n mantiene todas las filas del DataFrame original `df_grades`, independientemente de si hay coincidencias en el DataFrame `other_grades.csv`.

````python
right_merge = pd.merge(df_grades, df_other_grades, how='right', on='names')
display(right_merge)
````

El right merge conserva solo los elementos que se hayan combinado correctamente provenientes dataset `right`, en este caso `df_other_grades`. Este tipo de uni√≥n mantiene todas las filas del DataFrame other_grades.csv, independientemente de si hay coincidencias en el DataFrame original df_grades.

````python
inner_merge = pd.merge(df_grades, df_other_grades, how='inner', on='names')
display(inner_merge)
````

El inner join combina los elementos que se encuentren en ambas tablas. Descarta todo el resto. Este tipo de uni√≥n resulta en un DataFrame con menos filas que el DataFrame original `df_grades`, ya que solo incluye los registros que est√°n presentes en ambas tablas.

Basandonos en los resultados y la necesidad del rector de incluir los nuevos datos al dataset ya trabajado, sin tener que limpiarlos nuevamente, se deberia ocupar un innerjoin, de esta forma se obtendrian los alumnos con las notas que se encuentran en ambas tablas, esto indirectamente aplicaria los diferentes filtros y limpieza al dataset agregado, por otra parte si se debe unir al dataset original que no se le ha aplicado ningun cambio, es mejor hacer un outerjoin para conservar la mayor cantidad de datos ya que junta los datos de ambas tablas llenando con nan donde no coinciden y posteriormente hacer la limpieza correspondiente.
