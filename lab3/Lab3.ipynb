### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempeño, la Universidad les solicita que obtengan los alumnos quienes han reprobado el año y los que han tenido desempeño de destacado. Para esto se considerará que un alumno ha reprobado el año cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerará destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una función que permita obtener tres Dataframes:
- El primero deberá poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero será un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el año.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
  reprobados = dataframe_in[dataframe_in['GPA'] < 4.0]

  df_reprobados = dataframe_in.copy()
  df_reprobados.loc[df_reprobados.index.isin(reprobados.index), 'GPA'] = 'R'

  destacados = dataframe_in[(dataframe_in['math score'] > 5.8) & (dataframe_in['reading score'] > 5.8) & (dataframe_in['writing score'] > 5.8)].sort_values(by=['GPA'], ascending=False)

  reprobados = reprobados.reset_index(drop=True)
  destacados = destacados.reset_index(drop=True)

  return reprobados, df_reprobados, destacados
````

(investigar `qcut()`)

Hecho esto, grafique a través de un scatter plot el nivel educacional de los padres v/s el promedio de los alumnos, utilice los cuantiles generados anteriormente para mostrar en que cuantil se encuentra cada uno de los datos.

Realizado el grafico, responda la siguiente pregunta: **¿Es posible visualizar alguna tendencia en el gráfico?.**



**Respuesta:**

````python
# Código para quintiles


"""
NOTA: La variable del dataframe que contenga los quintiles debe llamarse
'GPA cuantiles' para que el gráfico funcione.
"""

def clasificar_quintiles(dataframe_in):
  dataframe_in['GPA cuantiles'] = pd.qcut(dataframe_in['GPA'], q=5, labels=['Bajísimo', 'Bajo', 'Medio', 'Alto', 'Muy alto'])
  return dataframe_in
````

### 4.2 Análisis por Nivel Educacional Etnia de los Padres [0.8 Puntos]

El rector, basado en su experiencia, cree fuertemente que el nivel educacional y la etnia de los padres influyen en las notas que obtienen sus hijos.
Como científicos de datos, ud. y su equipo creen que deben encontrar evidencia para confirmar o refutar la hipótesis del rector.

Para esto, deciden generar dos análisis: una tabla de resumen por una parte y gráficos de caja por otro.


**Tabla de Resumen [0.5 Puntos] **


Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión:

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.

Utilizar la tabla de resultados esperados como guía para desarrollar este punto.


**Análisis [0.3 Puntos]**

Finalmente, responda las siguientes preguntas sobre el gráfico de caja que viene implementado a continuación. En este se muestra el GPA con respecto al nivel educacional, donde la variable de color es la etnicidad.

- [ ]  ¿Hay alguna diferencia entre los grupos graficados tanto para el nivel educacional de los padres como también para la etnicidad?


````python
df_grades['GPA'] = df_grades[['math score', 'reading score', 'writing score']].mean(axis=1)

mapeo = {
    'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    "bachelor's degree": 'college',
    "associate's degree": 'college',
    "master's degree": 'postgraduate'
}

df_grades['simple parental level of education'] = df_grades['parental level of education'].map(mapeo)

tabla_resumen = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).agg({
    'math score': 'mean',
    'reading score': 'mean',
    'writing score': 'mean',
    'GPA': 'mean'
}).reset_index()

conteo_alumnos = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')

tabla_resumen = pd.merge(tabla_resumen, conteo_alumnos, on=['race/ethnicity', 'simple parental level of education'])

total_alumnos = df_grades.shape[0]
tabla_resumen['percentage'] = (tabla_resumen['count'] / total_alumnos * 100).round(2).astype(str) + ' %'

display(tabla_resumen)
````

````python
fig = px.box(
    df_grades,
    x="simple parental level of education",
    y="GPA",
    color="race/ethnicity",
    hover_data=["GPA"],
)
fig.show()
````

En la figura se puede apreciar un patrón en el GPA de las distintas razas/etnias que se repite entre niveles educacionales. Por ejemplo, siempre el grupo E es el que tiene el GPA promedio más alto entre niveles educacionales, mientras que el grupo A siempre tiene el GPA promedio más bajo entre estos. Por lo que la de los padres raza/etnia si influye en el desempeño de sus hijos.

Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situación (y así evitar buscar duplicados).

````python
df_other_grades =pd.read_csv('/content/drive/MyDrive/Lab programación cientifica/lab3/other_grades.csv')
display(df_other_grades)
df_grades =pd.read_json ('/content/drive/MyDrive/Lab programación cientifica/lab3/students_grades.json')
numeric_cols = ['math score', 'reading score', 'writing score']
df_grades[numeric_cols] = df_grades[numeric_cols].apply(pd.to_numeric, errors='coerce')
display(df_grades)
````

````python
outer_merge = pd.merge(df_grades, df_other_grades, how='outer', on='names')
display(outer_merge)
````

El outer merge combina todos los elementos posibles y conserva todo el resto en filas independientes.  Incluye todas las filas de ambos DataFrames, completando con NaNs donde no haya coincidencia entre las columnas 'names'.

````python
left_merge = pd.merge(df_grades, df_other_grades, how='left', on='names')
display(left_merge)
````

El left merge conserva solo los elementos que se hayan combinado correctamente provenientes dataset `left`, en este caso `df_grades`. Incluye todas las filas de `df_grades`, completando con NaNs donde no haya coincidencia entre las columnas 'name' en `other_grades.csv`. Este tipo de unión mantiene todas las filas del DataFrame original `df_grades`, independientemente de si hay coincidencias en el DataFrame `other_grades.csv`.

````python
right_merge = pd.merge(df_grades, df_other_grades, how='right', on='names')
display(right_merge)
````

El right merge conserva solo los elementos que se hayan combinado correctamente provenientes dataset `right`, en este caso `df_other_grades`. Este tipo de unión mantiene todas las filas del DataFrame other_grades.csv, independientemente de si hay coincidencias en el DataFrame original df_grades.

````python
inner_merge = pd.merge(df_grades, df_other_grades, how='inner', on='names')
display(inner_merge)
````

El inner join combina los elementos que se encuentren en ambas tablas. Descarta todo el resto. Este tipo de unión resulta en un DataFrame con menos filas que el DataFrame original `df_grades`, ya que solo incluye los registros que están presentes en ambas tablas.

Basandonos en los resultados y la necesidad del rector de incluir los nuevos datos al dataset ya trabajado, sin tener que limpiarlos nuevamente, se deberia ocupar un innerjoin, de esta forma se obtendrian los alumnos con las notas que se encuentran en ambas tablas, esto indirectamente aplicaria los diferentes filtros y limpieza al dataset agregado, por otra parte si se debe unir al dataset original que no se le ha aplicado ningun cambio, es mejor hacer un outerjoin para conservar la mayor cantidad de datos ya que junta los datos de ambas tablas llenando con nan donde no coinciden y posteriormente hacer la limpieza correspondiente.
